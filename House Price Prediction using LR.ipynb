```python
# Ml Libraries

import statsmodels.api as sma
from sklearn.linear_model import LinearRegression

from sklearn.model_selection import train_test_split

from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error,mean_absolute_percentage_error
```

```python
# OLS Model

x=  data.drop(columns = ['SalePrice'])
y = data['SalePrice']

base_model =  sma.OLS(y,sma.add_constant(x)).fit()
base_model.summary()
```

### Predictive Modeling

```python
# lets  split the dataset

x_train,x_test,y_train,y_test = train_test_split(x,y,train_size = 0.75, random_state=1)
```

```python
def model_scores(trainx,trainy,testx,testy,model):
    m = model
    m.fit(trainx,trainy)
    pred = m.predict(testx)
    print('R - square :',r2_score(testy,pred))
    print('MSE:',mean_squared_error(testy,pred))
    print('RMSE:',np.sqrt(mean_squared_error(testy,pred)))
    print('MAPE:',mean_absolute_percentage_error(testy,pred))
```

```python
# Test Data

model_scores(x_train,y_train,x_test,y_test,LinearRegression())
```

```python
# Train Dataset

model_scores(x_train,y_train,x_train,y_train,LinearRegression())
```

#### a. Forward Selection Method   : 

Algorithm :
1. It start from the null model
2. It adds the feature with the highest correlation with target in the model.
3. It picks the second column from the remaining dataset with the highest correlation.
4. It build the model adding the new feature to it.
5. This repeats until it has included all the columns.

User can pass no. of columns he wants to select or he can find no. of columns with best score . 


#### b. Backward Selection Method

Algorithm :
1. It starts with full model
2. It removes the column which is least significant( highest p_value)
3. It again fit some model with k-1 features
4. Again it will remove a column which is least significant 
5. It will repeat until all the columns are exhausted.

User can pass no. of columns he want to select or he can find no. of columns with best score.
